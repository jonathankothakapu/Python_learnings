```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

explore the official Matplotlib web page: http://matplotlib.org/  


```python
x=np.arange(0,10)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.ylim(0,20)
plt.title('Title1')

#plt.savefig('Plots/myfirstplot.jpg')
```

```python
#OOP -- layer by layer
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
a = np.linspace(0,11,10)
a
```

```python
b=a**4
b
```

```python
#plotting
fig = plt.figure(figsize=(3,3),dpi=100)  #set dpi for clarity (tradeoff with memory)
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel('a')
axes1.set_xlim(0,12)
axes1.set_title('X vs Y')

#Small axes
axes2 = fig.add_axes([0.1,0.4,0.5,0.5])
axes2.plot(x,y)
axes2.set_title('A vs B')

#fig.savefig('Plots/new_fig.png', bbox_inches='tight')
```

## Subplots

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_ylabel('Y label 0 0')

axes[0][1].plot(x,y**2)

axes[1][0].plot(a,b)

axes[1][1].plot(a,b+100)

fig.suptitle('SuperTitle', fontsize=20)

#plt.tight_layout() #automatically adjusts

fig.subplots_adjust(wspace=0.5,hspace=0.2)  #manually adjust
```

## Legend

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x')
ax.plot(x,x**2,label='x vs x^2')

#ax.legend(loc='center right')
#ax.legend(loc=(1.1,.5))
ax.legend(loc=(-.35,.5))
```

## Color and Markers

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='purple', label='x vs x', lw=3, ls='-.', marker='+',ms=20)
ax.plot(x,x*2, color='#3e9661', linewidth=5, linestyle=':', marker='p', markersize=30) #Hex color picker from Google

```

```python
# custom line
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
#ax.plot(x,x,color='purple', label='x vs x', lw=3, ls='-.')
lines = ax.plot(x,x*2, color='#3e9661', linewidth=2,marker='s',  ms=20,
               markerfacecolor='red',markeredgewidth=4 ,markeredgecolor='blue')


lines[0].set_dashes([5,3,10,3])   #5 points - 3 spaces - 10points- 3 spaces  -- (repeat)
```
