# Distribution Plots
1D  

For deviation, avgs, mode etc.   
Frequeny  
Rug plot  
Histogram  
Kernel Density Estimation (KDE) Plot. Ref: https://mathisonian.github.io/kde/  

```python
# Rugplot ( no meaning for y axis as this is just a distribution plot)
sns.rugplot(x='salary',data=df, height=0.8)
```

```python
# Distribution plot
# Histogram
#sns.distplot() is deprecated

sns.set(style='darkgrid')  #whitegrid, white, dark, ticks #sets for whole notebook
sns.displot(x='salary',data=df,bins=20)  #can change bin size
```

```python
# key word arguements available
sns.displot(x='salary',data=df,bins=20, color='black', edgecolor='red', linewidth=4, ls='--')
```

```python
#Rug + KDE+ Histogram
sns.displot(data=df,x='salary', kde=True,rug=True)
```

```python
#histplot
sns.histplot(data=df,y='salary')
```

```python
sns.histplot(data=df,x='salary')
```

```python
#Just KDE
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age',height=0.7)
```

```python
#bandwidth in Kernel and Bins here are analogous

sns.displot(data=sample_ages,x='age', rug=True, bins=30, kde=True)
```

```python
# Clip or range in KDE
#bandwidth is small, greater the gaussian dist plots (it is bandwidth of gaussian plot)
# adjust bandwidth properly to show the real dist of data
sns.kdeplot(data=sample_ages,x='age', clip=[0,100], bw_adjust=0.3, shade=True)
```

displot is better to use due to greater flexibility

## Distribution within Categories  
1. Boxplot  
2. Violin plot  
3. Swarmplot
4. Boxenplot (Letter-Value plot)  

```python
df1 = pd.read_csv('StudentsPerformance.csv')
```

```python
df1.head()
```

```python
sns.boxplot(data=df1, y='math score')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df1, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10,6), dpi=200)
sns.boxplot(data=df1, x='math score', y='parental level of education' ,hue='test preparation course')
plt.legend(bbox_to_anchor=(1.01,0.5))
```

**Violinplot**  
Violinplot is KDE mirrored. So it can adjusted by bandwidth of gaussian dist.

```python
plt.figure(figsize=(10,4))
sns.violinplot(data=df1, x='math score', y='parental level of education')  #you can add hue, palette etc.
```

```python
# Using hue parameter in split  --- not readable actually. Use this inly where it is needed.
plt.figure(figsize=(10,4))
sns.violinplot(data=df1, x='math score', y='parental level of education', 
               hue='test preparation course', split=True, inner='quartile', palette = 'Set1')
plt.legend(bbox_to_anchor=(1.2,0.5))

#lines inside violin show quartiles
# inner='stick' will show every instance of data
```

```python
# bandwidth 
plt.figure(figsize=(10,4))
sns.violinplot(data=df1, x='math score', y='parental level of education', bw=0.1)
```

### Swarm plot  

showing all the data points.

```python
sns.swarmplot(data=df1,x='math score')
```

```python
# If you get the above error, either increase figsize or decrease dot size. latter is better.
sns.swarmplot(data=df1,x='math score', size=3) 
```

```python
plt.figure(figsize=(8,4), dpi=100)
sns.swarmplot(data=df1,x='math score',y='gender', hue='test preparation course',size=3) 
plt.legend(bbox_to_anchor=(1.05,.5))
```

```python
#It is bad to read, we can use dodge=True
plt.figure(figsize=(8,4), dpi=100)
sns.swarmplot(data=df1,x='math score',y='gender', hue='test preparation course',size=3, dodge=True) 
plt.legend(bbox_to_anchor=(1.05,.5))
```

### Boxenplot  
Need more to understand. Audience may not understand.

```python
plt.figure(figsize=(8,3), dpi=200)
sns.boxenplot(data=df1, x='math score', y='test preparation course', hue='gender')
```

### Comparison Plots  
1. jointplot() -- scatterplot, hexagon, 2D KDE    
2. pairplot()

Warning      
pairplot() can be CPU and RAM intensive. Better to filter the data we are interested an dthen use pairplot(). 

```python
# jointplot()--  scatter plot, two continuous features used  
# maths score vs reading score

sns.jointplot(data=df1, x='math score', y='reading score')
```

```python
# using hexagon - draws an hexagon and more the points fall in the area of hex, more dark it is.
sns.jointplot(data=df1, x='math score', y='reading score',kind='hex')  #kind='scatter' by default
```

```python
sns.jointplot(data=df1, x='math score', y='reading score',kind='scatter', alpha=0.3)
```

```python
# kind='hist'
sns.jointplot(data=df1, x='math score', y='reading score',kind='hist')
```

```python
# 2D KDE
sns.jointplot(data=df1, x='math score', y='reading score',kind='kde', shade=True)
```

```python
# Hue
sns.jointplot(data=df1, x='math score', y='reading score', hue='gender')
```

```python
#goes through entire df and takes numerical varables and does comparison. So CPU and RAM intensive.
sns.pairplot(data=df1)
```

```python
sns.pairplot(data=df1, hue='gender', corner=True)  #corner eliminates duplicates
```

## Seaborn Subplots  
Create subplots based on grids  
Map plots based on pairplot()grid  

```python
sns.catplot(data=df1, x='gender', y='math score', kind='box', hue='lunch')
```

```python
sns.catplot(data=df1, x='gender', y='math score', kind='box', col='lunch')
```

```python
sns.catplot(data=df1, x='gender', y='math score', kind='violin', 
            col='lunch', row='test preparation course')
# careful what parameters to add. It can get complex easily.
```

```python
#leveraging PairGrid

g = sns.PairGrid(df1,hue='gender')
g = g.map_upper(sns.scatterplot)  #upper diagonal
g = g.map_diag(sns.kdeplot)    #along diagonal
g = g.map_lower(sns.kdeplot)   #lower diagonal
```

## Matrix Plots  

visual equivalent of Pivot table  

Heatmap  
Clustermap

```python
df2 = pd.read_csv('country_table.csv')
df2
#Countries column represent average of all the countries of the region.
```

```python
#Need a labaled index (categorical varable)
df2 = df2.set_index('Countries')
df2
```

```python
#assumes all the variables are in same units, compare only varables with same units.
sns.heatmap(df2.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
           cmap='viridis')
```

```python
sns.clustermap(df2.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
           cmap='viridis')  #col_cluster=False for  no col clustering
```
