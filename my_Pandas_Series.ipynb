## Useful methods -- .apply() on single column

**We want to get last four digits if CC number (Credit card number)
Steps:
1. Define a function (it should return only a single value, not a series of values)
2. use .apply(funtion) on the required column

```python
# quality of tips -- Generous or not
def quality(total_bill,tip):
    if tip/total_bill > .25:
        return 'Generous'
    else:
        return 'Other'
```

# Merge   

Inner Merge

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
#help(pd.merge)
```

```python
pd.merge(registrations,logins,how='inner',on='name')  #order does not matter for inner
```

## Left Join

```python
#shows everything in left table. If they don't have right table data, it shows NaN
pd.merge(left=registrations, right=logins,how='left',on='name')
#pd.merge(registrations,logins,how='left',on='name')   #both work
```

```python
## Right Join
pd.merge(left=registrations, right=logins,how='right',on='name')  #pandas smart enough join 'on' if it is same in both tables
```

## Outer Join   
Gives all the tables (union)   

```python
pd.merge(logins,registrations,how='outer',on='name')
```

```python
# join on column whose header is different in tables

#1st we set name in registrations as index
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
pd.merge(left=registrations,right=logins,left_index=True,right_on='name',how='inner')
```

```python
# reset the index
registrations = registrations.reset_index()
```

```python
registrations
```

```python
# renaming the columns
registrations.columns = ['reg_name','reg_id']
```

```python
registrations
```

```python
pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name')
```
